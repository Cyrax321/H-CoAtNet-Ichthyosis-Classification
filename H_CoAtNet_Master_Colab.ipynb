## Results Aggregation

```python
import os, glob
import numpy as np
from sklearn.metrics import accuracy_score, f1_score

PROJ = '/content/drive/MyDrive/HCoAtNet/H-CoAtNet-Ichthyosis-Classification/H-CoAtNet'

models = {
    'H-CoAtNet (Ours)':          ('h_coatnet_y_true.npy',                   'h_coatnet_y_pred.npy'),
    'BiomedCLIP (Fine-tuned)':   ('biomedclip_y_true.npy',                  'biomedclip_y_pred.npy'),
    'DINOv2-Base (Fine-tuned)':  ('dinov2_y_true.npy',                      'dinov2_y_pred.npy'),
    'EfficientNet-B0 (pretrained)':('efficientnet_pretrained_y_true.npy',   'efficientnet_pretrained_y_pred.npy'),
    'Swin-T (pretrained)':       ('swin_pretrained_y_true.npy',             'swin_pretrained_y_pred.npy'),
    'ViT-B/16 (pretrained)':     ('vit_pretrained_y_true.npy',              'vit_pretrained_y_pred.npy'),
    'CoAtNet (pretrained)':      ('coatnet_y_true.npy',                     'coatnet_y_pred.npy'),
    'GFT':                       ('gft_y_true.npy',                         'gft_y_pred.npy'),
    'CNN (scratch)':             ('cnn_y_true.npy',                         'cnn_y_pred.npy'),
    'EfficientNet-B0 (scratch)': ('efficientnet_scratch_y_true.npy',        'efficientnet_scratch_y_pred.npy'),
    'ViT (scratch)':             ('vit_scratch_y_true.npy',                 'vit_scratch_y_pred.npy'),
    'Swin-T (scratch)':          ('swin_scratch_y_true.npy',                'swin_scratch_y_pred.npy'),
}

print(f'{"Model":<35} {"Accuracy":>10} {"Macro F1":>10} {"Wtd F1":>10}')
print('-' * 68)
for name, (tf, pf) in models.items():
    tp = os.path.join(PROJ, tf)
    pp = os.path.join(PROJ, pf)
    if os.path.exists(tp) and os.path.exists(pp):
        yt = np.load(tp); yp = np.load(pp)
        acc = accuracy_score(yt, yp) * 100
        mf1 = f1_score(yt, yp, average='macro', zero_division=0)
        wf1 = f1_score(yt, yp, average='weighted', zero_division=0)
        print(f'{name:<35} {acc:>9.2f}% {mf1:>10.4f} {wf1:>10.4f}')
    else:
        print(f'{name:<35} {"NOT RUN YET":>10}')
```
